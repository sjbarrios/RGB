Desarrolle una función la cual reciba dos parámetros, una imagen y un entero llamado color, la función debe devolver una imagen la cual tenga activos los canales de color según los siguientes puntos:

- Si el parámetro color vale 1, la imagen debe mostrar activos únicamente el color azul.
- Si el parámetro color vale 2, la imagen debe mostrar activos únicamente el color verde.
- Si el parámetro color vale 3, la imagen debe mostrar activos únicamente el color rojo.
- Si el parámetro color vale 10, la imagen debe mostrar activos únicamente los colores rojo y verde.
- Si el parámetro color vale 20, la imagen debe mostrar activos únicamente los colores verde y azul.
- Si el parámetro color vale 30, la imagen debe mostrar activos únicamente los colores azul y rojo.

```python
image = cv2.imread('mercedes.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # OpenCV asume que el formato es BGR, por lo que se transforma a RGB
plt.imshow(image)
cv2.imwrite('./Problema1/original.jpg',cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
def filtro_imagen(imagen,color):
    
    imagen_filtrada = imagen.copy()
    
    #OpenCV asume que la imagen está en formato BGR (Blue-Green-Red)
    
    # Caso Azul
    if color == 1:
        #imagen_filtrada[:, :, 0] = 0    # Cancelación de Canal Azul
        imagen_filtrada[:, :, 1] = 0    # Cancelación de Canal Verde
        imagen_filtrada[:, :, 2] = 0    # Cancelación de Canal Rojo 
    
    # Caso Verde
    if color == 2:
        imagen_filtrada[:, :, 0] = 0    # Cancelación de Canal Azul
        #imagen_filtrada[:, :, 1] = 0    # Cancelación de Canal Verde
        imagen_filtrada[:, :, 2] = 0    # Cancelación de Canal Rojo 
    
    # Caso rojo
    if color == 3:
        imagen_filtrada[:, :, 0] = 0    # Cancelación de Canal Azul
        imagen_filtrada[:, :, 1] = 0    # Cancelación de Canal Verde
        #imagen_filtrada[:, :, 2] = 0    # Cancelación de Canal Rojo 
    
    # Caso rojo y verde
    if color == 10:
        imagen_filtrada[:, :, 0] = 0    # Cancelación de Canal Azul
        #imagen_filtrada[:, :, 1] = 0    # Cancelación de Canal Verde
        #imagen_filtrada[:, :, 2] = 0    # Cancelación de Canal Rojo
    
    
    # Caso verde y azul
    if color == 20:
        #imagen_filtrada[:, :, 0] = 0    # Cancelación de Canal Azul
        #imagen_filtrada[:, :, 1] = 0    # Cancelación de Canal Verde
        imagen_filtrada[:, :, 2] = 0    # Cancelación de Canal Rojo
        
    # Caso rojo y azul
    if color == 30:
        #imagen_filtrada[:, :, 0] = 0    # Cancelación de Canal Azul
        imagen_filtrada[:, :, 1] = 0    # Cancelación de Canal Verde
        #imagen_filtrada[:, :, 2] = 0    # Cancelación de Canal Rojo
    
    return imagen_filtrada
        
```

```python
def reconstruccion(gris_rojo,gris_verde,gris_azul):
    
    ancho = gris_rojo.shape[0]
    alto = gris_rojo.shape[1]
    profundidad = 3
    
    imagen_reconstruida = np.zeros((ancho,alto,profundidad))
    
    rojo = gris_rojo[:,:,0] # Extracción de la primera capa del rojo
    verde = gris_verde[:,:,0] # Extracción de la primera capa del verde
    azul = gris_azul[:,:,0] # Extracción de la primera capa del azul
    
    # Re-Creación del arreglo tri-dimensional
    imagen_reconstruida = np.stack((rojo,verde,azul),axis=2)
            
    plt.imshow(imagen_reconstruida)
    
    return imagen_reconstruida
    
```

### Problema 3:
Cree una función que dada una imagen cree una escala de grises en tres dimensiones, tome en cuenta que su función debe crear 3 imágenes como salida. Para entregar este ejercicio debe incluir una las imágenes que haya utilizado como prueba y el resultado de las misma, no puede utilizar la imagen del Problema No. 2.

```python
def deconstruccion(imagen_entrada):
    
    # Matriz de Subplots(fila,columnas)
    plt.figure()
    f, axarr = plt.subplots(2,2,figsize=(20,20)) 
    
    imagen = imagen_entrada.copy()

    # Dimensiones de la Imagen
    ancho = imagen.shape[0]
    alto = imagen.shape[1]
    
    # Escala de grises
    gris_rojo = np.zeros((ancho,alto,1),dtype=np.int32)
    gris_verde = np.zeros((ancho,alto,1),dtype=np.int32)
    gris_azul = np.zeros((ancho,alto,1),dtype=np.int32)

    for i in range(0,ancho):
        for j in range(0,alto):
            pixel = imagen[i,j] # BGR
            gris_rojo[i,j]= pixel[0]
            gris_verde[i,j]= pixel[1]
            gris_azul[i,j]= pixel[2]

    # Ensamble de Imágenes en Escala de Grises
    escala_gris_roja = np.concatenate((gris_rojo,gris_rojo,gris_rojo),axis=2)
    escala_gris_verde = np.concatenate((gris_verde,gris_verde,gris_verde),axis=2)
    escala_gris_azul = np.concatenate((gris_azul,gris_azul,gris_azul),axis=2)
    
    # Arreglo de Imágenes
    axarr[0][0].title.set_text('Imagen Original')
    axarr[0][0].imshow(imagen)
    axarr[0][1].title.set_text('Gris Rojo')
    axarr[0][1].imshow(escala_gris_roja)
    axarr[1][0].title.set_text('Gris Verde')
    axarr[1][0].imshow(escala_gris_verde)
    axarr[1][1].title.set_text('Gris Azul')
    axarr[1][1].imshow(escala_gris_azul)
    
    return escala_gris_roja,escala_gris_verde,escala_gris_azul
```

```python
gr,gv,ga=deconstruccion(image)
cv2.imwrite('./Problema3/original.jpg',cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
cv2.imwrite('./Problema3/escala_gris_roja.jpg',gr)
cv2.imwrite('./Problema3/escala_gris_verde.jpg',gv)
cv2.imwrite('./Problema3/escala_gris_azul.jpg',ga)
```

### Problema 4:
Cree una función que dada una imagen, muestre el histograma de cada canal de color y el de escala de grises (utilice un promedio aritmético para su escala de grises, no puede usar funciones de opencv), sus histogramas deben incluir una línea vertical la cual muestre el valor de la media de la distribución.

R. Para este problema, reutilizaremos la función que filtra canales de colores implementada con anterioridad.

```python
def histogramas_imagen(imagen_entrada):
    
    image = imagen_entrada.copy()
    ancho, alto, profundidad = image.shape
    
    # Arreglo Plano para Canal Rojo
    imagen_rojo = filtro_imagen(image,3)      
    imagen_rojo = imagen_rojo[:,:,2]
    imagen_rojo_plana = imagen_rojo.reshape((ancho * alto * 1, 1))
    
    # Arreglo Plano para Canal Verde
    imagen_verde = filtro_imagen(image,2)
    imagen_verde = imagen_verde[:,:,1]
    imagen_verde_plana = imagen_verde.reshape((ancho * alto * 1, 1))
       
    # Arreglo Plano para Canal Azul
    imagen_azul = filtro_imagen(image,1)
    imagen_azul = imagen_azul[:,:,0]
    imagen_azul_plana = imagen_azul.reshape((ancho * alto * 1, 1))
    
    
    # Arreglo de Imágenes
    plt.figure()
    #subplot(fila,columnas)
    f, axarr = plt.subplots(1,4,figsize=(20,5)) 
    
    # Histograma Rojo
    media_rojo = np.mean(imagen_rojo_plana)
    axarr[0].title.set_text('Histograma Rojo')
    axarr[0].hist(imagen_rojo_plana, bins='auto', color='red')
    axarr[0].axvline(x=media_rojo, color='black', linestyle='dashed', linewidth=1)
    
    # Histograma Verde
    media_verde = np.mean(imagen_verde_plana)
    axarr[1].title.set_text('Histograma Verde')
    axarr[1].hist(imagen_verde_plana, bins='auto', color='green')
    axarr[1].axvline(x=media_verde, color='black', linestyle='dashed', linewidth=1)
    
    # Histograma Azul
    media_azul = np.mean(imagen_azul_plana)
    axarr[2].title.set_text('Histograma Azul')
    axarr[2].hist(imagen_azul_plana, bins='auto', color='blue')
    axarr[2].axvline(x=media_azul, color='black', linestyle='dashed', linewidth=1)
  
    # Escala de Grises (Promedio Aritmético)
    imagen_gris = np.zeros((ancho,alto,1))
    
    for i in range(0,ancho):
        for j in range(0,alto):
            pixel = image[i,j] # RGB
            prom_aritmetico = int((int(pixel[0]) + int(pixel[1]) + int(pixel[2]))/3)
            imagen_gris[i,j] = prom_aritmetico
    
    imagen_gris_plana = imagen_gris.reshape((ancho * alto * 1, 1))
    
    media_gris = np.mean(imagen_gris_plana)
    axarr[3].title.set_text('Histograma Escala Grises')
    axarr[3].hist(imagen_gris_plana, bins='auto', color='gray')
    axarr[3].axvline(x=media_gris, color='black', linestyle='dashed', linewidth=1)
    
    plt.show()
   
    
    return 0
```

### Problema 5:
Investigue en que consiste el enfoque de escala de grises ponderado, luego de esto implemente una función que dada una imagen, realice una escala de grises ponderada (notar que no existe una solución única).

### Escala de Grises Ponderada (Método de Escala de Grises Ponderado o Método de la Luminosidad)

Este método trata de corregir el potencial problema que ocurre con el método del **Promedio Aritmético** para Escala de Grises, en el cual se toma el promedio de los tres colores. El inconveniente es que los trés colores tienen tres diferentes *longitudes de onda* y por lo tanto, cada uno tiene una contribución diferente a la formación de la imagen. Por lo tanto, al promediar 33%, 33%, 33% no se está tomando en cuenta esta "diferencia de contribuciones". El método de la **luminosidad** intenta ponderar esta contibución de cada color a la formación de la imagen, basado en que cada color tiene distintas longitudes de onda, y por lo tanto es percibido de distinta manera para el ojo humano.

El color **rojo** tiene mayor longitud de onda que el **verde** y éste último mayor longitud de onda que el **azul**. El color verde tiene un efecto más perceptible para el ojo humano. Intuitivamente entonces, se debería reducir la contribución del color rojo, incrementar la contribución del color verde y ponderar en el medio al color azul.

```python
def escala_grises_ponderada(imagen_entrada):
    
    image = imagen_entrada.copy()
    ancho = image.shape[0]
    alto = image.shape[1]

    gris_ponderado = np.zeros((ancho,alto,1),dtype=np.int32)

    for i in range(0,ancho):
        for j in range(0,alto):
            pixel = image[i,j] # RGB
            promedio_ponderado = (int(0.299*pixel[0])+int(0.587*pixel[1])+int(0.114*pixel[2]))
            gris_ponderado[i,j] = int(promedio_ponderado)
    
    # Ensamble de Imágenes en Escala de Grises
    escala_gris_ponderado = np.concatenate((gris_ponderado,gris_ponderado,gris_ponderado),axis=2)
    
    # Matriz de Subplots(fila,columnas)
    plt.figure()
    f, axarr = plt.subplots(1,2,figsize=(20,20)) 
    
    axarr[0].title.set_text('Imagen Original')
    axarr[0].imshow(image)
    axarr[1].title.set_text('Escala de Grises Ponderada')
    axarr[1].imshow(escala_gris_ponderado)
     
    return escala_gris_ponderado
    
```

```python
gris_ponderado=escala_grises_ponderada(image)
cv2.imwrite('./Problema5/escala_gris_ponderada.jpg',gris_ponderado)
```
